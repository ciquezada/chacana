```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.visualization import SqrtStretch, simple_norm
from astropy.visualization.mpl_normalize import ImageNormalize
```

```python
IMAGE_X0 = 730
IMAGE_Y0 = 450
```

```python
def show_fit(image):
    norm = ImageNormalize(stretch=SqrtStretch())
    fig, ax = plt.subplots(figsize=(15,10))
    ax.set_title("ifts")
    im = ax.imshow(image, origin='lower', norm=norm)
    fig.colorbar(im)
    plt.legend()
    plt.show()
    plt.close()
```

```python
def masking(image):
    image_x0 = IMAGE_X0
    image_y0 = IMAGE_Y0
    image_masked = np.zeros((1040, 1548), dtype = "int16")
    image_masked = image.copy()
    for y in range(len(image)):
        for x in range(len(image[0])):
            if (x-image_x0)**2+(y-image_y0)**2>=450**2:
#                 image_masked[y][x] = image[y][x]
                image_masked[y][x] = 0
                
    return image_masked
```

```python
def extract_center(image):
    out = []
    image_x0 = IMAGE_X0
    image_y0 = IMAGE_Y0
    for y in range(len(image)):
        for x in range(len(image[0])):
            if (x-image_x0)**2+(y-image_y0)**2<=450**2:
                out.append(image[y][x])
    return np.array(out, dtype="int16")
```

## Probamos el metodo del profe

### primero la diferencia absoluta

```python
image1_diff = np.abs(image1_next - image1_prev)
show_fit(image1_next)
show_fit(masking(image1_diff))
```

### Aplicando suavizado a la diferencia (comparamos con suavizar primero y luego restar)

```python
from astropy.convolution import Gaussian2DKernel, convolve


kernel = Gaussian2DKernel(x_stddev=1)
image1_smoothed = convolve(masking(image1_diff), kernel)
show_fit(image1_smoothed)
image1_diff_smoothed = np.abs( convolve(masking(image1_next), kernel) - convolve(masking(image1_prev), kernel) )
show_fit(image1_diff_smoothed)
```
